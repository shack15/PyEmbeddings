```python
import pyembeddings
from pyembeddings import Generator, Database
```

```python
pyembeddings.init('e39b5850-078b-4738-94ce-9c273617b4c2')
gen = Generator()
db = Database()

sample_text = "Hello World"
embedding = gen.embed(sample_text)
print("Embedding:", embedding)
```

```python
print(gen.count_tokens("Hello World"))
print(gen.within_token_limit("Hello World"))
print(gen.get_model_info())
```

```python
gen.set_model("bge-base")
print(gen.get_model_info())
print(gen.count_tokens("A fairly complicated and long piece of convoluted text."))
sample_text = "A fairly complicated and long piece of convoluted text."
embedding = gen.embed(sample_text)
print(embedding)

gen.set_model("MiniLM")
print(gen.count_tokens(sample_text))
embedding = gen.embed(sample_text)
print(embedding)
```
